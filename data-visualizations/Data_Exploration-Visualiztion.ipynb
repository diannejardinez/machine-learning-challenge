### This Jupyter Notebook is used to visualize some of the data from the NASA Kelper space telescope raw dataset 

```python
import pandas as pd

import numpy as np
```

```python
# Load and Clean Data

df = pd.read_csv("exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()


# Set features(X) and target(y) values
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]
```

```python
# Show columns
df.columns

# Description of columns from NASA Exoplanet Archive
# Source Link: https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html
```

```python
# Correlation of some of the dataset- Heatmap
# Source Link: https://seaborn.pydata.org/generated/seaborn.heatmap.html

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
rcParams['figure.figsize'] = 14, 12

data = df.corr()
ax = sns.heatmap(data, cmap="YlGnBu", linewidths=.5)
```

#### Histograms

```python
# Confirmed planets and their Transit Properties Stellar Parameters
# Source Link: https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from matplotlib import rcParams
rcParams['figure.figsize'] = 10, 8

ax = df[df.koi_disposition == 'CONFIRMED'].koi_kepmag.hist(bins=100, figsize=(12, 8))
ax.set_xlabel("Kepler-band (mag)")
ax.set_title("Brightness of an object as it appears in the night sky")

```

```python

ax = df[df.koi_disposition == 'CONFIRMED'].koi_steff.hist(bins=100, figsize=(12, 8))
ax.set_xlabel("Stellar Effective Temperature (Kelvin))")
ax.set_title("The Star photospheric temperature")
```

```python

ax = df[df.koi_disposition == 'CONFIRMED'].koi_slogg.hist(bins=100, figsize=(12, 8))
ax.set_xlabel("Stellar Surface Gravity (log10(cm s-2)")
ax.set_title("The base-10 logarithm of the acceleration due to gravity at the surface of the star")
```

```python

ax = df[df.koi_disposition == 'CONFIRMED'].koi_prad.hist(bins=100, figsize=(12, 8))
ax.set_xlabel("Planetary Radius (Earth radii)")
ax.set_title("Planet Radius")
```
